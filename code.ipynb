# model building

```python
x=df.drop(columns=['Target'])
y=df['Target']
```

```python
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
x
```

```python
y
```

```python
#!pip install xgboost
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from imblearn.over_sampling import ADASYN

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
# handel imbalence using adasyn
adasyn=ADASYN(random_state=42)
xres,yres=adasyn.fit_resample(x,y)
xres.shape
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(xres,yres,test_size=0.2,random_state=42) 
```

```python
models={
    'Logisticregression':LogisticRegression(),
    'Random forest': RandomForestClassifier(n_estimators=200,random_state=42),
    'XGboost':XGBClassifier(random_state=42),
    'decision tree':DecisionTreeClassifier(random_state=42),
    'svm':SVC()
}
#model.items() # returns the pair of key:value in tuple fromat from dictionry
```

```python
for name,model in models.items():
    print(f'\ntraining{name}...')
    model.fit(xtrain,ytrain)

    train_acc=accuracy_score(ytrain,model.predict(xtrain))
    test_acc=accuracy_score(ytest,model.predict(xtest))

    print(f'{name}-Train_accuracy:{train_acc:.4f},test accuracy:{test_acc:.4f}')
```
